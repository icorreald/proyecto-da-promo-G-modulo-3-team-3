```python
#Contingencia, Chi Cuadrado. 

# ¿Está asociada la satisfacción en el entorno laboral de forma ignificativa con las relaciones interpersonales? 
# p_value < 0,05 rechazamos H0.Existe diferencia.
# p_value > 0,05 aceptamos H0 , no existe diferencia.  ✅

tabla_contingencia = pd.crosstab(df['environment_satisfaction'], df['relationship_satisfaction'])
chi2, p, _, _ = stats.chi2_contingency(tabla_contingencia)
print("Estadística chi-cuadrado:", chi2)
print("Valor p:", p)
```
