## COSAS HECHAS:

```python
# nombres columnas
columnas = df.columns.tolist()
columnas = columnas[1:]
columnas
```

```python
serie_columnas = pd.Series(columnas)
serie_columnas
```

```python
import re
```

```python
patron_may_min = re.compile(r'([a-z])([A-Z])')
patron_employee = re.compile(r'(employee)([A-Z][a-z]*)')

columnas_2 = []

for nombre in columnas:
    if patron_may_min.search(nombre):
        nombre = patron_may_min.sub(r'\1_\2', nombre)
        columnas_2.append(nombre.lower())

    elif patron_employee.search(nombre):
        nombre = patron_employee.sub(r'\1_\2', nombre)
        columnas_2.append(nombre)

    elif nombre == 'NUMCOMPANIESWORKED':
        nombre = 'num_companies_worked'
        columnas_2.append(nombre)

    elif nombre == 'Over18':
        nombre = 'over_18'
        columnas_2.append(nombre)

    else:
        columnas_2.append(nombre.lower())

columnas_2
        
```

```python
print(len(columnas))
print(len(columnas_2))
```
